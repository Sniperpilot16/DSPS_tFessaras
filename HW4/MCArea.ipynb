```
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
%pylab inline
```

```
def square(side):
  ax = plt.figure(figsize = (7,7)).add_subplot(1,1,1) 
  plt.scatter(-side,-side)
  plt.scatter(-side,side)
  plt.scatter(side,-side)
  plt.scatter(side,side)
  plt.plot([-side,-side],[-side,side])
  plt.plot([side,-side],[-side,-side])
  plt.plot([-side,side],[side,side])
  plt.plot([side,side],[side,-side])
  
  return ax

```

```
from astropy import units as u
```

```
def circle(radius,ax):
  angle = np.arange(0,360,0.1)
  angle *= u.degree
  x = radius * np.cos(angle.to(u.rad))
  y = radius * np.sin(angle.to(u.rad))
  print(x)
  print(y)
  
  ax.plot(x,y)
ax = plt.figure().add_subplot(1,1,1)
```

```
axis = square(1)
circle(1,axis)

np.random.seed(300)
num_X = (np.random.rand(10000) * 2)-1
num_Y = (np.random.rand(10000) * 2)-1

axis.scatter(num_X,num_Y)

r = np.sqrt((num_X**2) + (num_Y**2))
#print(r)
r_in = r[r<1]

print(len(r_in))
print((4*len(r_in))/10000)
```
